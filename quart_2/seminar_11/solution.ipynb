```python
# Условие задачи

Даны две функции: f(x) = x^3 - 50x и g(x) = -x^4 + 88x^2 - 241
Требуется:
1. Найти координаты точек пересечения
2. Построить графики функций в одной системе координат
3. Построить график функции пересечения
```

```python
from sympy import *
x = Symbol('x')
fx = x**3 - 50*x
gx = -1*x**4 + 88*x**2 - 241
fgx = fx - gx
```

```python
# Корни

roots = solve(fgx)
roots
```

```python
roots[1]
```

```python
# Словарь с x и y

cross_dict = {}
c = 1
for z in roots:
    cross_dict[c] = {'x': round(N(z), 5), 'y': round(N(fx.evalf(subs={x: z}, chop=True)), 5)}
    c += 1
cross_dict
```

```python
# Вывод координат точек пересечения

print(*[(cross_dict[i]['x'], cross_dict[i]['y']) for i in cross_dict.keys()], sep=', ')

```

```python
# Вывод графиков функий в одной системе координат

plot(fx, gx, x)
```

```python
# График функции пересечения

p = plot(fgx, x)
```

```python
p = plot(fx, gx, fgx, show=false)
p[2].line_color = 'g'
p.show()
```
